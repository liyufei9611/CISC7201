# Further Process the Data

```python
from ast import literal_eval
import itertools
import re
```

## Create sub-sheet for better understanding

* supported_languages
* platforms 
* tags 

### Create sub-sheet of 'language'

```python
# language_data['languages'].sample(20)
#English Russian Chinese Japanese Spanish
language_data = comment[['appid','name','languages']].copy()

language_data['is_English']=language_data['languages'].str.contains('English')
language_data['is_Chinese']=language_data['languages'].str.contains('Chinese')
language_data['is_Japanese']=language_data['languages'].str.contains('Japanese')
language_data['is_Russian']=language_data['languages'].str.contains('Russian')
language_data['is_Spanish']=language_data['languages'].str.contains('Spanish')

language_data.drop('languages',axis=1,inplace=True)
language_data.head(5)
```

### Create sub-sheet of 'tags'

```python
def parse_export_tags(x):
    x = literal_eval(x)
    if isinstance(x, dict):
        return x
    elif isinstance(x, list):
        return {}
    else:
        raise TypeError('Something other than dict or list found')
        
tag_data = comment[['appid','name','tags']].copy()
tag_data['tags'] = tag_data['tags'].apply(parse_export_tags)
cols_tags = set(itertools.chain(*tag_data['tags']))    #找出所有的tags

for col in sorted(cols_tags):
    col_name = col.lower().replace(' ', '_').replace('-', '_').replace("'", "")
    tag_data[col_name] = tag_data['tags'].apply(lambda x: x[col] if col in x.keys() else 0)

tag_data = tag_data.drop('tags', axis=1)
tag_data.head(5)
```

### Create sub-sheet of 'platforms'

```python
platforms_data = df[['steam_appid', 'platforms']].copy()

platforms_data['sup_windows'] = platforms_data['platforms'].str.contains("'windows': True")
platforms_data['sup_mac'] = platforms_data['platforms'].str.contains("'mac': True")
platforms_data['sup_linux'] = platforms_data['platforms'].str.contains("'linux': True")

platforms_data.drop('platforms', axis=1, inplace=True)
platforms_data.head(5)
```

### Convert 'release_date' to time series data

```python
release_date_data = df[['steam_appid', 'release_date']]
# ---------------------------------
def eval_date(x):
    x = literal_eval(x)
    if x['coming_soon']:
        return '' # return blank string so can drop missing at end
    else:
        return x['date']
# ---------------------------------    
def parse_date(x):
    if re.search(r'[\d]{1,2} [A-Za-z]{3}, [\d]{4}', x):
        return x.replace(',', '')
    elif re.search(r'[A-Za-z]{3} [\d]{4}', x):
        return '1 ' + x
    elif x == '':
        return np.nan
    else:
        # Should be everything, print out anything left just in case
        #print(x)
        pass
# ---------------------------------
def process_release_date(release_date_data):
    release_date_data = release_date_data[release_date_data['release_date'].notnull()]
    release_date_data = release_date_data.copy()
    

    release_date_data['release_date'] = release_date_data['release_date'].apply(eval_date)
    

    release_date_data['release_date'] = release_date_data['release_date'].apply(parse_date)
    release_date_data['release_date'] = pd.to_datetime(release_date_data['release_date'], format='%d %b %Y', errors='coerce')
    release_date_data = release_date_data[release_date_data['release_date'].notnull()]
    return release_date_data
release_date_data = process_release_date(release_date_data)
release_date_data.sample(5)
```

## Deal with Text 

* recommendations 处理后转int
* achievement 只读取成就总数，取int
* owners 处理后，保留左右值，原值存为平均数
* <font color = red>尚未处理developers / publishers 两个都是list形式，没想好

```python
df['recommendations'] = df['recommendations'].fillna("{'total':0}").str[9:-1].values
#df['recommendations'].sample(5)
```

```python
df['achievements'] = df['achievements'].str.replace("{'total': 0}", "{'total': 0,}").fillna("{'total': 0,").str.replace("{'total': ", "")
df['achievements'] = df['achievements'].str.split(",", n = 1, expand = True).values
```

```python
df[['owner_left', 'owner_right']] = df['owners'].str.split(' .. ', expand = True)

df['owner_left'] = df['owner_left'].str.replace(',', '')
df['owner_right'] = df['owner_right'].str.replace(',', '')

df['owner_left'] = pd.to_numeric(df['owner_left'], downcast = 'integer', errors='coerce').fillna(0)
df['owner_right'] = pd.to_numeric(df['owner_right'], downcast = 'integer', errors='coerce').fillna(0)

df['owners'] = (df['owner_left'] + df['owner_right'])  / 2
```
